```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/diabetes.csv')
df.head(10)
```

### Faltantes (nulos)

### Análisis estadístico

```python
df.describe()
```

### Gráficos de densidad

```python
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[col], fill=True, color='skyblue', bw_adjust=0.5)
    plt.title(f'Gráfico de densidad para  {col}')
    plt.tight_layout()
    plt.show()
```

### Boxplots

```python
for col in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df["Outcome"], y=df[col])
    plt.title(f'Boxplot de {col}')
    plt.xlabel("Outcome")
    plt.ylabel(col)
    plt.show()
```
